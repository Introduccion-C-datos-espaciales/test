```python
from IPython.display import IFrame  
ciaLink2="https://www.cia.gov/the-world-factbook/field/revenue-from-forest-resources/country-comparison" 
IFrame(ciaLink2, width=900, height=900)
```

```python
# read web table into pandas DF
import pandas as pd

forestDFs=pd.read_html(ciaLink2, # link
                        header=0, # where is the header?
                        flavor='bs4')
```

```python
# check object type 
type(forestDFs)
```

```python
# check size
len(forestDFs)
```

```python
# make a copy
forest=forestDFs[0].copy()
```

```python
# here it is
forest
```

```python
# see column names:

forest.columns
```

```python
# or

forest.columns.to_list()
```

```python
forest['Date of Information'].value_counts()
```

```python
#1 Replace '%' by 'pct'.
forest.columns=forest.columns.str.replace(pat=r'%',repl='pct',regex=True)
forest
```

```python
#2 Keep the columns Country, pct of GDP, and Date of Informa
forest=forest.drop(columns=["Rank"])
forest
```

```python
#3 Change the column name Date of Information to forest_date
forest=forest.rename(columns={"Date of Information": "forest_date"})
forest
```

```python
#4 Make sure there are no spaces as part of the column names
forest.columns=forest.columns.str.replace(pat=r' ',repl='',regex=True)
forest
```

```python
#5 Make sure the cells with text does not have neither trailing nor leading spaces
forest['Country']=forest['Country'].str.strip()
forest
```

```python
#6 Keep only the year value in the column forest_date
forest.forest_date.str.extract(pat=r'(\d+)',expand=True)
```
